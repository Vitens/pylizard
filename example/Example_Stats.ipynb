```python
import pylizard, matplotlib.pyplot as plt, numpy
```

```python
code, filt = '16FL0041', 1
p = pylizard.Peilbuis(code, filt, report=True)
```

```python
p.plot(figsize=[16, 9]);
```

```python
h_mean = p.head_hand.resample('1D').mean()
h_linear = p.head_hand.resample('1D').bfill(limit=1).interpolate()
h_poly = p.head_hand.resample('1D').bfill(limit=1).interpolate('polynomial', order=2)
```

```python
fig, ax = plt.subplots(figsize=[16, 9])
h_linear.plot(ax=ax, style='b-')
h_poly.plot(ax=ax, style='g-')
h_mean.plot(style='r.', ms=10, ax=ax)
```
