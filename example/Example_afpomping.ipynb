```python
import pylizard, matplotlib.pyplot as plt, pastas
%matplotlib inline
```

```python
code, filt = '39DP0222', 2
p = pylizard.Peilbuis(code, filt)
```

```python
p.head_diver.plot()
```

```python
h_min = p.head_diver.resample('D').min()
h_max = p.head_diver.resample('D').max()
h_dif = h_max-h_min
```

```python
h_min.rolling(7).mean().plot()
h_max.rolling(7).mean().plot()
```

```python
fig, ax = plt.subplots(figsize=[16, 9])
h_dif.plot(ax=ax)
h_dif.rolling(30).mean().plot(ax=ax)
ax.set_ylim(0, 5)
plt.show()
```
