```python
import pylizard, matplotlib.pyplot as plt
%matplotlib inline
```

```python
code, filt = '39DP0222', 2
p = pylizard.Peilbuis(code, filt)
```

Standard plot

```python
p.head_diver.plot(figsize=[16, 9]);
```

Resample the head to day and determine the maximum en minimum head per day. The maximim head is an approximate for the head with no pumping and the head minimum vice versa. So the difference is approximately the drawdown at the monitoring filter.

```python
h_1day = p.head_diver.resample('D')
h_min = h_1day.min()
h_max = h_1day.max()
h_dif = h_max-h_min
```

Plot maximum and minimum head per day

```python
fig, ax = plt.subplots(figsize=[16, 9])
h_min.rolling(7).mean().plot(ax=ax)
h_max.rolling(7).mean().plot(ax=ax)
ax.legend(['min' ,'max'])
```

Plot the approximate drawdown and the 30 day rolling average

```python
fig, ax = plt.subplots(figsize=[16, 9])
h_dif.plot(ax=ax)
h_dif.rolling(30).mean().plot(ax=ax)
ax.legend(['difference', '30 day rolling average'])
```
