Collect monitoring well within distance from point

```python
x, y, dist = 153000, 475450, 5000
buis = pylizard.pnt2buis(x, y, dist)
buis
```

Plot topview and x, y crosssections of the filters

```python
fig = plt.figure(figsize=[16, 9])
grid = plt.GridSpec(5, 5, hspace=0.5, wspace=0.4)

main_ax = fig.add_subplot(grid[:-1, 1:], aspect='equal', adjustable='datalim')

y_ax = fig.add_subplot(grid[:-1, 0])
x_ax = fig.add_subplot(grid[-1, 1:])

main_ax.plot(buis['x'], buis['y'], 'r.', ms=10)
x_ax.plot(buis['x'], buis['okf'], 'r.', ms=10)
y_ax.plot(buis['okf'], buis['y'], 'r.', ms=10)

main_ax.apply_aspect()
xlim = main_ax.get_xlim()
ylim = main_ax.get_ylim()
y_ax.set_ylim(ylim)
x_ax.set_xlim(xlim)

main_ax.set_title('Bovenaanzicht',fontweight='bold')
y_ax.set_title('Dwarsdoorsnede y-as',fontweight='bold')
x_ax.set_title('Dwarsdoorsnede x-as',fontweight='bold');
```

Filter only with measurements

```python
buis_meting = buis.loc[(buis['uuid_diver']!='') | (buis['uuid_hand']!='')]
buis_meting
```

Plot all timeseries

```python
fig, ax = plt.subplots(figsize=[16, 9])
label = []
for r in buis_meting.iterrows():
    print(r[1]['buis'], r[1]['filt'])
    p = pylizard.Peilbuis(r[1]['buis'], r[1]['filt'])
    if r[1]['uuid_hand']!='' and r[1]['uuid_diver']!='': 
        p.plot(ax=ax, stats=False)
    elif r[1]['uuid_hand']!='':
        p.head_hand.plot(ax=ax)
    elif r[1]['uuid_diver']!='':
        p.head_diver.plot(ax=ax)
    label.append(r[0])
ax.legend(label)
ax.set_ylabel('Stijghoogte (m+N.A.P.)')
ax.set_xlabel('Datum')
ax.set_title('Peilbuizen binnen {}m van {}, {}'.format(dist, x, y));
```
